```python
import os
```

```python
directory = '/media/rajlab/DATASETS/ADNI_BIDS'
adni_folders = []
# iterate over files in
# that directory
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    # checking if it is a file
    if os.path.isdir(f):
        adni_folders.append(f)
```
